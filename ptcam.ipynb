```python
import os
from PIL import Image

# Path to image folder
image_folder = "D:/Kiran/PTCAM11_stitch/ptcam11_images"
image_files = [f for f in os.listdir(image_folder) if f.startswith("PTCAM11_") and f.endswith(".bmp")]

# Sort files by number to ensure they are in the correct order
image_files.sort(key=lambda x: int(x.split('_')[1].split('.')[0]))

# Separate even and odd files
odd_images = [f for f in image_files if int(f.split('_')[1].split('.')[0]) % 2 != 0]
even_images = [f for f in image_files if int(f.split('_')[1].split('.')[0]) % 2 == 0]

# Define overlap size (number of pixels to overlap at borders)
overlap_size = 4  # Adjust this if needed

# Helper function to stitch images with overlap
def stitch_images_with_overlap(image_files, output_path, overlap_size):
    # Load first image to get dimensions
    first_image = Image.open(os.path.join(image_folder, image_files[0]))
    image_width, image_height = first_image.size

    # Calculate the final stitched image height considering overlap
    stitched_image_height = (image_height - overlap_size) * len(image_files) + overlap_size
    stitched_image = Image.new('RGB', (image_width, stitched_image_height))

    # Initialize the y_offset for placement in the stitched image
    y_offset = 0

    for image_file in image_files:
        img = Image.open(os.path.join(image_folder, image_file))

        # Calculate the area to paste, trimming overlap at the top except for the first image
        cropped_img = img.crop((0, 0, img.width, img.height))

        # Paste cropped image at current y_offset
        stitched_image.paste(cropped_img, (0, y_offset))

        # Increment y_offset by the height minus overlap for the next image
        y_offset += image_height - overlap_size

    # Save the final stitched image
    stitched_image.save(output_path)
    print(f"Stitching completed with overlap. Image saved as '{output_path}'.")

# Stitch odd-numbered images into one image with overlap
stitch_images_with_overlap(odd_images, "stitched_image_odd_with_overlap.bmp", overlap_size)

# Stitch even-numbered images into another image with overlap
stitch_images_with_overlap(even_images, "stitched_image_even_with_overlap.bmp", overlap_size)
```
